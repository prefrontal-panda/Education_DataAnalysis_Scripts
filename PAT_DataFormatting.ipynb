# Cleaning PAT data

```python
# Import libraries
import pandas as pd
```

```python
# Importing files
data = pd.read_csv("C:/Users/Debbie.Chong/Virtual_env/PAT2025_IDC.csv")
```

```python
data.head()
```

```python
index = ['Family name', 'Given name', 'Middle names', 'Gender', 
         'Date of birth', 'Year level (at time of test)', 'Username']

data = data.pivot(index = index, columns='Test', values='Scale score')
```

```python
# Removing multi-indexing
data = data.reset_index()
```

```python
data.info()
```

```python
# Rearranging columns
# Start and end columns
col_ord_start = ['Given name', 'Middle names', 'Family name']
cold_ord_end = ['Gender', 'Date of birth', 'Year level (at time of test)', 'Username'] #'PAT Maths Adaptive', 

# Getting test columns
cols = data.columns.tolist()
tests = [col for col in cols if col.startswith('Test')]
tests = sorted(tests, key=lambda x: int(x.split(' ')[1]))

# Rearranging
data = data[col_ord_start + tests + cold_ord_end]
```

```python
# Getting rid of special characters in name columns
data['Middle names'] = data['Middle names'].str.replace('\xa0', '', regex=False)
```

```python
# Saving
data.to_csv('PAT2025_Maths_SEQTAUpload_IDC.csv', index=False, header=True, encoding='utf-8')
```
